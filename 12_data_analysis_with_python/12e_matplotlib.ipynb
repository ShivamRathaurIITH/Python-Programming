Matplotlib is a library, Pyplot is a module (or submodule) within the Matplotlib library.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# create a line plot
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# create a customized line plot
plt.plot(x,y, color='green', linestyle=':', marker='o', linewidth='2', markersize='8')
plt.grid(True, tickdir='in')
plt.show()
```

```python
# multiple plots
x = [1, 2, 3, 4, 5]
y1 = [11, 24, 39, 16, 25]
y2 = [10, 20, 30, 40, 50]
y3 = [121,99,155,33,110]

# plt.figure() creates a new plotting area where you can plot your data.
plt.figure(figsize=(10,4)) # create a figure of size 10x4 inches # (width,height) -> for overall plot not for sub-plots

plt.subplot(1,3,1)
plt.plot(x,y1,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot I')

plt.subplot(1,3,2)
plt.plot(x,y2,color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot II')

plt.subplot(1,3,3)
plt.plot(x,y3,color='purple')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot III')

# Show the plots
plt.tight_layout()  # Adjust subplots to avoid overlap
plt.show()
```

### Bar Plot

```python
## bar plot
cate = ['A','B','C','D','E']
val = [5,7,3,8,6]

# create a bar plot
plt.bar(cate,val,color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')

plt.show()
```

### Histogram
* Histograms are used to represent the distribution of a dataset. 
* They divide the data into bins and count the number of data points in each bin.

```python
# sample data
data = [13, 1, 7, 1, 6, 6, 2, 11, 9, 13, 4, 2, 9, 1, 14, 15, 9, 15, 7, 5, 4]

# create a histogram
plt.hist(data,bins=8,color='red',edgecolor='black')
plt.xlabel('Intervals')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

```python
# scatter plot
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.scatter(x,y,color='red', marker='x')

plt.show()
```

### Pie Chart

```python
# pie chart
sizes = [30,20,40,10]
lbs = ['A','B','C',"D"]
clr = ['red','yellow','purple','green']
exp = (0.1,0,0,0.05) # "pull out" or "separate" one or more slices from the rest of the chart.

plt.pie(sizes, labels=lbs, colors=clr, explode=exp, autopct='%1.1f%%', shadow=True) 

plt.show()
```

```python
# autopct: "automatic percentage" (for each slice)
'''
String format: A format string for the percentage.
    Example: '%1.1f%%' displays percentages with one decimal place.
    Example: '%1.0f%%' displays percentages as whole numbers.

'%1.1f%%':
    The 1 before the decimal ensures at least 1 digit before the decimal point (but allows more digits if the percentage exceeds 9.9%).
    The .1f ensures exactly 1 digit after the decimal.
    %%: Displays the % symbol.    
'''
```
